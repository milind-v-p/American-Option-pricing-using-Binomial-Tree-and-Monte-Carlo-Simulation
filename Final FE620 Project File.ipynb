```python
import numpy as np
import pandas as pd
from scipy.stats import norm
import yfinance as yf
import matplotlib.pyplot as plt

ticker = 'NVDA'
start_date = '2023-04-01'
end_date = '2024-04-16'
stock_data = yf.download(ticker, start=start_date, end=end_date)['Close']
log_returns = np.log(stock_data / stock_data.shift(1)).dropna()
mhu = log_returns.mean() * 252
sigma = log_returns.std() * np.sqrt(252)
#print(sigma)

# Download Treasury yield data
treasury_data = yf.download('^TNX', start='2024-01-01', end='2024-12-31')

# Select the most recent Treasury yield value (risk-free rate)
risk_free_rate = treasury_data['Close'].iloc[-1] / 100
#print("Risk-Free Rate:", risk_free_rate)
S0 = stock_data.iloc[-1]
print("S0 = ", S0)

def black_scholes(S, K, T, r, sigma, option_type):
    """
    Calculate European option price using the Black-Scholes formula.

    Parameters:
    S (float): Stock price
    K (float): Strike price
    T (float): Time to expiration in years
    r (float): Risk-free interest rate (annual)
    sigma (float): Volatility of the underlying (annual)

    Returns:
    float: Option price
    """
    # Calculate d1 and d2
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)

    if option_type == 'call':
        price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    elif option_type == 'put':
        price = K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
    
    return price

def check_put_call_parity(call_prices, put_prices, S, K_values, r_annual, days_to_expiration, days_in_year):
    T = days_to_expiration / days_in_year
    parity_checks = []

    for i, K in enumerate(K_values):
        C = call_prices.loc[call_prices['Strike Price'] == K, 'Option Price(BT)'].iloc[0]
        P = put_prices.loc[put_prices['Strike Price'] == K, 'Option Price(BT)'].iloc[0]
        parity = C - P
        expected = S - K * np.exp(-r_annual * T)
        parity_checks.append({
            "Strike Price": K,
            "C - P": parity,
            "S - Ke^(-rT)": expected,
            "Parity Difference": round((-parity + expected),5)
        })

    return pd.DataFrame(parity_checks)

def generate_gbm(S, r_annual, sigma_annual, days_to_expiration, days_in_year, steps, n):
    delta_t = days_to_expiration / days_in_year / steps
    gbm_values = np.zeros((n, steps + 1))
    
    for i in range(n):
        gbm_values[i, 0] = S
        for j in range(1, steps + 1):
            z = np.random.normal(0, 1)
            gbm_values[i, j] = gbm_values[i, j - 1] * np.exp((r_annual - 0.5 * sigma_annual**2) * delta_t + sigma_annual * np.sqrt(delta_t) * z)
    
    average_gbm_values = np.mean(gbm_values, axis=0)
    
    return average_gbm_values

def calculate_option_price(S, K, r_annual, sigma_annual, days_to_expiration, days_in_year, steps, n, payoff="call"):
    delta_t = days_to_expiration / days_in_year / steps
    u = np.exp(sigma_annual * np.sqrt(delta_t))
    d = 1 / u
    p = (np.exp(r_annual * delta_t) - d) / (u - d)
    
    average_gbm_values = generate_gbm(S, r_annual, sigma_annual, days_to_expiration, days_in_year, steps, n)
    
    option_values = np.zeros(steps + 1)
    
    for j in range(steps + 1):
        if payoff == "call":
            option_values[j] = max(0, average_gbm_values[j] - K)
        elif payoff == "put":
            option_values[j] = max(0, K - average_gbm_values[j])
    
    # Threshold for negligible option value
    threshold = 0.001 * max(S - K, K - S)  # 1% of the maximum intrinsic value
    
    # Check if option is deeply out of the money and set value to zero if below threshold
    if np.all(option_values < threshold):
        return 0
    
    for j in range(steps - 1, -1, -1):
        for i in range(j + 1):
            option_values[i] = np.exp(-r_annual * delta_t) * (p * option_values[i] + (1 - p) * option_values[i + 1])
            exercise_value = max(0, average_gbm_values[i] - K) if payoff == "call" else max(0, K - average_gbm_values[i])
            option_values[i] = max(option_values[i], exercise_value)
    
    return option_values[0]

def calculate_greeks(S, K, T, r, sigma, option_type):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))

    delta = norm.cdf(d1) if option_type == 'call' else norm.cdf(d1) - 1
    gamma = norm.pdf(d1) / (S * sigma * np.sqrt(T))

    return delta, gamma

# Parameters
S = S0
K_values = [690, 700, 710, 720, 730, 740, 750, 760, 770, 780, 790, 800, 805, 810, 815, 820, 825, 830, 835, 840, 845, 850, 
            855, 860, 865, 870, 875, 880, 885, 890, 895, 900, 905, 910, 920]
r_annual = risk_free_rate  # Annual risk-free rate
sigma_annual = log_returns.std() * np.sqrt(252)  # Annualized volatility
days_to_expiration = 24
days_in_year = 252
steps = 150
n=10^6
T = days_to_expiration / days_in_year  # Time to expiration in years
payoff = "call"
# Results DataFrames
results_call = []
results_put = []

for K in K_values:
    price_call = black_scholes(S, K, T, r_annual, sigma_annual, 'call')
    price_put = black_scholes(S, K, T, r_annual, sigma_annual, 'put')
    price_call_binomial_tree = calculate_option_price(S, K, r_annual, sigma_annual, days_to_expiration, days_in_year, steps,n, payoff = "call")
    price_put_binomial_tree = calculate_option_price(S, K, r_annual, sigma_annual, days_to_expiration, days_in_year, steps,n, payoff = "put")
    delta_call, gamma_call = calculate_greeks(S, K, T, r_annual, sigma_annual, 'call')
    delta_put, gamma_put = calculate_greeks(S, K, T, r_annual, sigma_annual, 'put')
    
    results_call.append({'Strike Price': K, 'Option Price(BSM)': price_call,'Option Price(BT)': price_call_binomial_tree, 'Delta': delta_call, 'Gamma': gamma_call})
    results_put.append({'Strike Price': K, 'Option Price': price_put,'Option Price(BT)': price_put_binomial_tree, 'Delta': delta_put, 'Gamma': gamma_put})

df_call = pd.DataFrame(results_call)
df_put = pd.DataFrame(results_put)

# print("Call Option Prices and Greeks:")
# print(df_call)
# print("\nPut Option Prices and Greeks:")
# print(df_put)

parity_table = check_put_call_parity(df_call, df_put, S, K_values, r_annual, days_to_expiration, days_in_year)
print("\nPut-Call Parity Check for Binomial Tree Method:")
print(parity_table)

# plt.figure(figsize=(10, 6))
# plt.plot(df_call['Strike Price'], df_call['Option Price(BSM)'], label='Option Price (BSM) - Call', marker='o')
# plt.plot(df_call['Strike Price'], df_call['Option Price(BT)'], label='Option Price (BT) - Call', marker='o')
# plt.plot(df_call['Strike Price'], df_call['Delta'], label='Delta - Call', marker='o')
# plt.plot(df_call['Strike Price'], df_call['Gamma'], label='Gamma - Call', marker='o')
# plt.title('Call Option Metrics vs Strike Price')
# plt.xlabel('Strike Price')
# plt.ylabel('Value')
# plt.legend()
# plt.grid(True)
# plt.xticks(df_call['Strike Price'])
# plt.xticks(df_call['Strike Price'][::4]) 
# plt.show()

# # Plot for Put options
# plt.figure(figsize=(10, 6))
# plt.plot(df_put['Strike Price'], df_put['Option Price'], label='Option Price - Put', marker='o')
# plt.plot(df_put['Strike Price'], df_put['Option Price(BT)'], label='Option Price (BT) - Put', marker='o')
# plt.plot(df_put['Strike Price'], df_put['Delta'], label='Delta - Put', marker='o')
# plt.plot(df_put['Strike Price'], df_put['Gamma'], label='Gamma - Put', marker='o')
# plt.title('Put Option Metrics vs Strike Price')
# plt.xlabel('Strike Price')
# plt.ylabel('Value')
# plt.legend()
# plt.grid(True)
# plt.xticks(df_put['Strike Price'])
# plt.xticks(df_call['Strike Price'][::4]) 
# plt.show()

plt.figure(figsize=(8,5))
plt.plot(df_call['Strike Price'], df_call['Option Price(BSM)'], label='Option Price (BSM) - Call', marker='o')
plt.title('Black Scholes Prices vs Strike Price')
plt.xlabel('Strike Price')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.xticks(df_call['Strike Price'][::4])
plt.tight_layout()
plt.show()

# Plot for Call options
plt.figure(figsize=(8,14.4))

# Plot for Call option prices
plt.subplot(3, 1, 1)
plt.plot(df_call['Strike Price'], df_call['Option Price(BSM)'], label='Option Price (BSM) - Call', marker='o')
plt.plot(df_call['Strike Price'], df_call['Option Price(BT)'], label='Option Price (BT) - Call', marker='o')
plt.title('Call Option Prices vs Strike Price')
plt.xlabel('Strike Price')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.xticks(df_call['Strike Price'][::4])

# Plot for Call option delta
plt.subplot(3, 1, 2)
plt.plot(df_call['Strike Price'], df_call['Delta'], label='Delta - Call', marker='o')
plt.title('Delta vs Strike Price (Call)')
plt.xlabel('Strike Price')
plt.ylabel('Delta')
plt.legend()
plt.grid(True)
plt.xticks(df_call['Strike Price'][::4])

# Plot for Call option gamma
plt.subplot(3, 1, 3)
plt.plot(df_call['Strike Price'], df_call['Gamma'], label='Gamma - Call', marker='o')
plt.title('Gamma vs Strike Price (Call)')
plt.xlabel('Strike Price')
plt.ylabel('Gamma')
plt.legend()
plt.grid(True)
plt.xticks(df_call['Strike Price'][::4])

plt.tight_layout()
plt.show()

# Plot for Put options
plt.figure(figsize=(8,14.4))

# Plot for Put option prices
plt.subplot(3, 1, 1)
plt.plot(df_put['Strike Price'], df_put['Option Price'], label='Option Price - Put', marker='o')
plt.plot(df_put['Strike Price'], df_put['Option Price(BT)'], label='Option Price (BT) - Put', marker='o')
plt.title('Put Option Prices vs Strike Price')
plt.xlabel('Strike Price')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.xticks(df_put['Strike Price'][::4])

# Plot for Put option delta
plt.subplot(3, 1, 2)
plt.plot(df_put['Strike Price'], df_put['Delta'], label='Delta - Put', marker='o')
plt.title('Delta vs Strike Price (Put)')
plt.xlabel('Strike Price')
plt.ylabel('Delta')
plt.legend()
plt.grid(True)
plt.xticks(df_put['Strike Price'][::4])

# Plot for Put option gamma
plt.subplot(3, 1, 3)
plt.plot(df_put['Strike Price'], df_put['Gamma'], label='Gamma - Put', marker='o')
plt.title('Gamma vs Strike Price (Put)')
plt.xlabel('Strike Price')
plt.ylabel('Gamma')
plt.legend()
plt.grid(True)
plt.xticks(df_put['Strike Price'][::4])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import yfinance as yf
import matplotlib.pyplot as plt

ticker = 'NVDA'
start_date = '2023-04-01'
end_date = '2024-04-16'
stock_data = yf.download(ticker, start=start_date, end=end_date)['Close']
log_returns = np.log(stock_data / stock_data.shift(1)).dropna()
mhu = log_returns.mean() * 252
sigma = log_returns.std() * np.sqrt(252)
#print(sigma)

# Download Treasury yield data
treasury_data = yf.download('^TNX', start='2024-01-01', end='2024-12-31')

# Select the most recent Treasury yield value (risk-free rate)
risk_free_rate = treasury_data['Close'].iloc[-1] / 100
#print("Risk-Free Rate:", risk_free_rate)
S0 = stock_data.iloc[-1]


def calculate_option_price(S, K, r_annual, sigma_annual, days_to_expiration, days_in_year, steps, payoff = "call"):
    delta_t = days_to_expiration / days_in_year / steps
    u = np.exp(sigma_annual * np.sqrt(delta_t))
    d = 1 / u
    p = (np.exp(r_annual * delta_t) - d) / (u - d)

    stock_prices = np.zeros(steps + 1)
    option_values = np.zeros(steps + 1)

    for i in range(steps + 1):
        stock_prices[i] = S * (u**(steps - i)) * (d**i)
        if payoff == "call":
            option_values[i] = max(0, stock_prices[i] - K)
        elif payoff == "put":
            option_values[i] = max(0, -stock_prices[i] + K)

    for j in range(steps - 1, -1, -1):
        for i in range(j + 1):
            option_values[i] = np.exp(-r_annual * delta_t) * (p * option_values[i] + (1 - p) * option_values[i + 1])

    return option_values[0]

def calculate_sensitivity(S, K, r_annual, sigma_monthly, days_to_expiration, days_in_year, steps, dS):
    base_price = calculate_option_price(S, K, r_annual, sigma_monthly, days_to_expiration, days_in_year, steps,n)
    price_up = calculate_option_price(S + dS, K, r_annual, sigma_monthly, days_to_expiration, days_in_year, steps,n)
    price_down = calculate_option_price(S - dS, K, r_annual, sigma_monthly, days_to_expiration, days_in_year, steps,n)

    delta = (price_up - price_down) / (2 * dS)
    gamma = (price_up - 2 * base_price + price_down) / (dS**2)

    return base_price, delta, gamma

S = 840.35  # Current stock price
K = 840  # Strike price
r_annual = 0.04633  # Annual risk-free rate
days_to_expiration = 29
days_in_year = 252
sigma_monthly = sigma / np.sqrt(252) * np.sqrt(12)  # Monthly volatility
dS = S * 0.01  # Small percentage change in stock price
n=10^5

# Analyze sensitivity to number of steps
steps_range = range(2, 150)  # Steps from 2 to 100
results = []

for steps in steps_range:
    price, delta, gamma = calculate_sensitivity(S, K, r_annual, sigma_monthly, days_to_expiration, days_in_year, steps, dS)
    results.append({
        'Steps': steps,
        'Option Price': price,
        'Delta': delta,
        'Gamma': gamma
    })

results_df = pd.DataFrame(results)

# Plotting
plt.figure(figsize=(14, 8))
plt.subplot(3, 1, 1)
sns.lineplot(data=results_df, x='Steps', y='Option Price', marker='o', label='Option Price')
plt.title('Option Price Sensitivity to Number of Steps')
plt.xlabel('Number of Steps')
plt.ylabel('Option Price')

plt.subplot(3, 1, 2)
sns.lineplot(data=results_df, x='Steps', y='Delta', marker='o', color='green', label='Delta')
plt.title('Delta Sensitivity to Number of Steps')
plt.xlabel('Number of Steps')
plt.ylabel('Delta')



plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Download Treasury yield data
treasury_data = yf.download('^TNX', start='2024-01-01', end='2024-12-31')
risk_free_rate = treasury_data['Close'].iloc[-1] / 100

# Function to calculate the Black-Scholes delta
def black_scholes_delta(S, K, T, r, sigma, option_type='call'):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    if option_type == 'call':
        delta = norm.cdf(d1)
    else:
        delta = norm.cdf(d1) - 1
    return delta

# Function to execute trades based on delta and update position
def execute_trades(delta, current_position, initial_position, option_quantity):
    trade_quantity = delta * option_quantity - current_position
    current_position += trade_quantity
    return current_position, trade_quantity

# Load stock data
ticker = 'NVDA'
start_date = '2024-04-01'
end_date = '2024-04-17'
stock_data = yf.download(ticker, start=start_date, end=end_date)['Close']

# Option details
strike_price = 900
option_quantity = 1000

# Estimate parameters
log_returns = np.log(stock_data / stock_data.shift(1)).dropna()
mu = log_returns.mean() * 252
sigma = log_returns.std() * np.sqrt(252)
S0 = stock_data.iloc[0]
T = (np.datetime64('2024-04-17') - np.datetime64('2024-04-01')).astype(int) / 252
r = risk_free_rate  # Placeholder value for risk-free rate

# Initialize position
initial_position = 500  # Initial stock position
current_position = initial_position

# Lists to store positions and dates
positions = [initial_position]
dates = [np.datetime64('2024-04-01')]

# Main loop for delta hedging
for index, row in stock_data.iloc[:-1].items():
    S0 = row
    delta = black_scholes_delta(S0, strike_price, T, r, sigma)
    current_position, trade_quantity = execute_trades(delta, current_position, initial_position, option_quantity)
    positions.append(current_position)
    dates.append(index)

# Plotting positions over time
plt.figure(figsize=(10, 6))
plt.plot(dates, positions, marker='o', linestyle='-')
plt.title('Delta Hedging Positions Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Stocks Held')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
